```python
import pandas as pd
```

```python
file = 'Resources/Tables/23173_table.csv'
```

```python
census_df = pd.read_csv(file, encoding='ISO-8859-1')
census_df.head()
```

```python
drop_col = census_df.drop(['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
drop_col.head()
```

```python
dropna = drop_col.dropna(how='any')
dropna.head()
```

```python
df = dropna.rename(columns={"Subject": "Subject", "ZCTA5 23173": "Estimate", 'Unnamed: 5': 'Margin of Error', \
                           'Unnamed: 6': 'Percent', 'Unnamed: 7': 'Percent Margin of Error'})
df.head()
```
